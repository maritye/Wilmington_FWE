```python
dfs = [RX1DW, RX5DW, PTOTW, SDIIW, R10mmW, R20mmW, CDDW, CWDW]
indices = ['RX1D', 'RX5D', 'PTOT', 'SDII', 'R10mm', 'R20mm','CDD', 'CWD']
```

```python
for i, df in enumerate(dfs):
    print(indices[i])
    cur = (df.sel(year=slice('1981-01-01','2010-12-31')).mean('year').mean('proj')).to_array(name='CUR')
    r45= (df.isel(proj=0).sel(year=slice('2040-01-01','2069-12-31')).mean('year')).to_array(name = 'Fut_r45')
    r85 = (df.isel(proj=1).sel(year=slice('2040-01-01','2069-12-31')).mean('year')).to_array(name='Fut_r85')

    new = xr.merge([cur, r45, r85], compat='override')
    newnm = iandoDir + 'LOCA_CESM1.' + indices[i] + '.climatologies.DJF.nc'
    new.to_netcdf(newnm)
    
    

    fig, axs = plt.subplots(1,3, figsize = (12,5))

    cur.plot(ax=axs[0])
    r45.plot(ax=axs[1])
    r85.plot(ax=axs[2])

#    pltnm = indices[i] + '.draftplots.DJF.jpg'
#    plt.savefig(pltnm)

    plt.show();
```
